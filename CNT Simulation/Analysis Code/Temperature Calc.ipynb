```python
import numpy as np 
import matplotlib.pyplot as plt


I_list = np.linspace(1,10,num=10000) # mA
# I_list = np.array([0.1, 1])
E = 1161 # J
C = 9e-3 # F
f = 60 # Hz
V = 500 # volts

Z = 1 / 2 / np.pi / f / C # Ohms
P_list = (I_list/1000)**2 * Z # Watts
t_list = C * V / (I_list/1000) # sec
t_list_min = t_list / 60 # min
m_list = np.array([6.17e-8, 1.74e-10, 1.48e-8, 3.06e-9, 9e-8, 5e-12, 1.7e-12, 2.52e-11, 2.87e-12, 8.65e-11, 8.65e-11, 1.5e-8]) # kg
c_list = np.array([120, 730, 710, 120, 500, 462, 887, 753, 540, 730, 730, 710]) # J / kg /degC
mc_list = m_list * c_list # J / degC
mc_total = np.sum(mc_list) # J / degC
Q_list = t_list * P_list # Joules
temp_list = Q_list / mc_total

plt.figure(figsize=(8,6))
plt.plot(t_list_min, temp_list)
plt.title('Temperature Change Assuming No Heat Dissipation', fontsize=20)
plt.xlabel('Charging Time (min)', fontsize=16)
plt.ylabel(R'Temperature Change ($^{\circ}$C)', fontsize=16)
plt.savefig(R'D:\Local Documents\Research\Khalid\Temperature\Temp_vs_current.png',dpi=300,bbox_inches='tight')
plt.show()
```
